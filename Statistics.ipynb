## Soal Pertama (10 poin)

Tunjukkan statistika deskriptif untuk variabel harga rumah 
(Column Name : SalePrice)!

Ceritakan kepada tim bisnis apa makna dari nilai nilai 
tersebut!

**Ekspektasi output :**

● ukuran pemusatan (4 poin)

● ukuran penyebaran (4 poin)

● Interpretasi (2 poin)

```python
import pandas as pd
import numpy as np
import scipy.stats as st

import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
house_price_v = pd.read_csv('train.csv')
house_price_v.head()
```

```python
# Descriptive Statistics
house_price_v.describe().transpose()[['mean','50%']]
```

```python
sale_price = house_price_v['SalePrice']

# Measure of Tendency
print('rata-rata: ',sale_price.mean())
print('median: ',sale_price.median())
print('modus:',sale_price.mode()[0])

#Quartile
print('Q1',sale_price.quantile(0.25)) #Q1
print('Q2',sale_price.quantile(0.50)) #Q2
print('Q3',sale_price.quantile(0.75)) #Q3
```

```python
# Measure of Spread

# Variance
print('Variance:',sale_price.var())
# Standard Deviation
print('Standard Deviation:',sale_price.std())
# Inter Quartile Range
print('Inter Quartile Range (IQR):',sale_price.quantile(0.75)-sale_price.quantile(0.25))
# Range
print('Range:',sale_price.max()-sale_price.min())
```

**Interpretasi**

Dari hasil yang ditampilkan, dapat dilihat bahwa rata-rata harga rumah adalah sekitar 180,921.20. Namun, karena median (163,000.00) lebih kecil dari rata-rata, ini menunjukkan bahwa distribusi harga rumah mungkin skew ke kanan. Modus (140,000) adalah harga rumah yang paling sering muncul dalam data.

Kuartil menunjukkan bagaimana data dibagi menjadi empat bagian yang sama. Q1 (129,975.00) adalah nilai yang membagi 25% data terendah, sedangkan Q3 (214,000.00) adalah nilai yang membagi 25% data tertinggi. Inter Quartile Range (IQR) adalah jarak antara Q1 dan Q3 dan menunjukkan variasi harga rumah di tengah 50% data.

Variance menunjukkan seberapa jauh nilai-nilai data tersebar dari rata-rata. Standard Deviation adalah akar kuadrat dari variance dan menunjukkan variasi harga rumah dalam satuan yang sama dengan data. Range (720,100) adalah selisih antara nilai tertinggi dan terendah dalam data dan menunjukkan variasi total harga rumah.

```python
house_price_year = pd.read_csv('train.csv')
house_price_year.head()
```

```python
house_price_year['YrSold'].value_counts()
```

```python
price_year = house_price_year.groupby('YrSold')['SalePrice'].describe()
price_year = price_year.transpose()
price_year.transpose()[['mean','50%','25%','75%','std']]
```

```python
df_6 = house_price_year[house_price_year['YrSold'] == 2006]
df_7 = house_price_year[house_price_year['YrSold'] == 2007]
df_8 = house_price_year[house_price_year['YrSold'] == 2008]
df_9 = house_price_year[house_price_year['YrSold'] == 2009]
df_10 = house_price_year[house_price_year['YrSold'] == 2010]

ftest = st.f_oneway(df_6['SalePrice'],
                   df_7['SalePrice'],
                   df_8['SalePrice'],
                   df_9['SalePrice'],
                   df_10['SalePrice'])
p_value = ftest.pvalue
alpha = 0.05
print('P-value', p_value)
```

```python
sns.boxplot(x='YrSold', y='SalePrice', data=house_price_year)
```

**Interpretasi**

Dari tabel yang dihasilkan, terlihat bahwa rata-rata harga rumah (mean) bervariasi setiap tahunnya, dengan harga tertinggi pada tahun 2007 dan terendah pada tahun 2008. Namun, perbedaan ini tidak terlalu signifikan. Jika dilihat dari ukuran penyebaran (std), variasi harga rumah juga bervariasi setiap tahunnya, dengan variasi tertinggi pada tahun 2007 dan terendah pada tahun 2008. Secara keseluruhan, tidak ada perbedaan yang signifikan dalam harga rumah setiap tahunnya.

**Kesimpulan**

```python
from scipy.stats import mannwhitneyu
import numpy as np
```

```python
campaign_y = pd.read_csv('bank-full.csv', engine='python', sep=None)
campaign_y.head()
```

```python
N = campaign_y.shape[0]
e = 0.05
slovin = N / (1+N*e**2)
slovin
```

**Penentuan Hipotesis**

H0 : jumlah campaign tidak memiliki efek pada keinginan user untuk membuka akun deposit <br>
H1 : jumlah campaign memiliki efek positif pada keinginan user untuk membuka akun deposit

```python
yes = campaign_y[campaign_y['y']=='yes']
no = campaign_y[campaign_y['y']=='no']
```

```python
yes.shape
```

```python
no.shape
```

```python
apply_yes = yes['campaign'].sum()
apply_no = no['campaign'].sum()

obs_yes = yes['campaign'].count()
obs_no = no['campaign'].count()
```

```python
amount_yes = apply_yes/obs_yes
amount_no = apply_no/obs_no
```

```python
print('Amount from yes : {0:0.4f}'.format(amount_yes))
print('Amount from no : {0:0.4f}'.format(amount_no))
```

```python
x = ['yes','no']
value = [amount_yes,amount_no]
plt.bar(x,value)

plt.ylim(2.0,3.0)
```

```python
apply = (apply_yes,apply_no)
obs = (obs_yes,obs_no)
```

```python
stats, pvalue = mannwhitneyu(apply,obs)
print('P-value',pvalue)
```

```python
if pvalue >= 0.05:
    print('Terima H0 artinya jumlah campaign tidak memiliki efek pada keinginan user untuk membuka akun deposit')
else:
    print('Terima H1 artinya jumlah campaign memiliki efek positif pada keinginan user untuk membuka akun deposit')
```

**Interpretasi**

Dari data yang dihasilkan menunjukkan bahwa nilai p-value yang lebih besar dari 0.05. Ini berarti kita tidak dapat menolak hipotesis nol (H0), dan secara grafik 'no' lebih tinggi daripada 'yes'.

**Kesimpulan**

Kesimpulannya adalah bahwa jumlah campaign tidak memiliki efek yang signifikan pada keinginan user untuk membuka akun deposit.
